```python
import numpy as np
import scipy.linalg as la
from scipy import stats
import scipy
import time
```

```python
import matplotlib.pyplot as plt
import matplotlib
font = {'family' : 'serif',
        'size'   : 13}

matplotlib.rc('font', **font)
matplotlib.rc('xtick', labelsize=15) 
matplotlib.rc('ytick', labelsize=15) 
```

```python
k1 = 2*1e3
k2 = 2*1e-3
k3 = 22
CB0 = 0.5
CA0 = 5*1e-5
CD0 = 0
CP0 = 0
x0 = [CB0, CA0, CD0, CP0]
```

```python
def jacobian_matrix(x):
    CB, CA, CD, CP = x
    return np.array([[-k1*CA, -k1*CB, k2, 0], 
                     [-k1*CA, -k1*CB, k2 + k3, 0], 
                     [k1*CA, k1*CB, -k2 - k3, 0], 
                     [0, 0, k3, 0]])
```

```python
A0 = jacobian_matrix(x0)
```

```python
np.linalg.cond(A0)
```

```python
E, V = np.linalg.eig(A0)
```

```python
E
```

```python
plt.plot(np.arange(1, len(E)+1), np.flip(np.sort(np.abs(E))), "-o")
plt.yscale("log")
plt.ylabel("$|\lambda|$")
plt.xlabel("Index")
plt.title("Eignevalues of the Jacobian Matrix")
plt.tight_layout()
plt.savefig("figs2/Jacobian_Eigs_p2.png", dpi=400)
```

```python
def fdot(t, x):
    CB, CA, CD, CP = x
    return np.array([-k1*CB*CA + k2*CD, -k1*CB*CA + (k2+k3)*CD, k1*CB*CA - (k2+k3)*CD, k3*CD])
```

```python
t0 = time.time()
res = scipy.integrate.solve_ivp(fun=fdot, 
                          t_span=(0, 1000), 
                          y0=x0,
                          method='BDF')
t1 = time.time()
print(t1-t0)
print(res)
print(len(res.t))
```

```python
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(10 , 6))
ax[0][0].plot(res.t, res.y[0])
ax[0][1].plot(res.t, res.y[1])
ax[1][0].plot(res.t, res.y[2])
ax[1][1].plot(res.t, res.y[3])


ax[0][0].set_title("$C_{B}$")
ax[0][1].set_title("$C_{A}$")
ax[1][0].set_title("$C_{D}$")
ax[1][1].set_title("$C_{P}$")

ax[0][0].set_xlabel("t")
ax[1][0].set_xlabel("t")
ax[0][1].set_xlabel("t")
ax[1][1].set_xlabel("t")


plt.tight_layout()
plt.savefig("figs2/BDF_chemical_solution.png", dpi=400)
```

```python
def rk4(y, h):
    k1 = h* fdot(0, y)
    k2 = h*fdot(0, y+0.5*k1)
    k3 = h*fdot(0, y+0.5*k2)
    k4 = h*fdot(0, y+k3)
    return y + 1/6*k1 + 1/3*(k2 + k3) + 1/6*k4
```

```python
t0 = time.time()
h = 0.001 
t = np.linspace(0, 1000, int(1000/h) +1)
y_res = np.zeros((len(x0), len(t)))
y_res[:, 0] = x0

for ii, t_s in enumerate(t[:-1]):
    y_res[:, ii + 1] = rk4(y_res[:, ii], h)

t1 = time.time()
res.t = t
res.y = y_res
```

```python
def falken_skan(t, y):
    y1, y2, y3 = y
    return np.array([-y3*y1 - beta*(1-y2**2), y1, y2])
```

```python
m = -0.09
beta = 2*m/(m+1)
```

```python
beta
```

```python
res1 = scipy.integrate.solve_ivp(fun=falken_skan, 
                      t_span=(0, 10), 
                      y0=np.array([1, 0, 0]),
                      method='RK45')
res2 = scipy.integrate.solve_ivp(fun=falken_skan, 
                      t_span=(0, 10), 
                      y0=np.array([0, 0, 0]),
                      method='RK45')
eps= np.inf
ii = 0

y1_array = [1, 0]

z_np = 1
z_n = 0
while eps >= 1E-4:
    z_new = z_n - (res2.y[1][-1] - 1)*(z_n - z_np)/(res2.y[1][-1] -res1.y[1][-1] )
    z_np = np.copy(z_n)
    z_n = np.copy(z_new)
    
    y1_array.append(z_n)
    res1 = scipy.integrate.solve_ivp(fun=falken_skan, 
                      t_span=(0, 10), 
                      y0=np.array([z_np, 0, 0]),
                      method='RK45')
    res2 = scipy.integrate.solve_ivp(fun=falken_skan, 
                          t_span=(0, 10), 
                          y0=np.array([z_n, 0, 0]),
                          method='RK45')

    eps = np.abs(res2.y[1][-1] - 1)
    

    ii += 1
```
