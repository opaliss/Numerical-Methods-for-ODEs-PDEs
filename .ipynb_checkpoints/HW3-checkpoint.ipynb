# Problem 2

# (a)
$$
 \frac{dT_{j}}{dt} + u \frac{T_{j+1
}- T_{j-1}}{2\Delta x} = 0
$$

CFL Condition: 
$$
\Delta t < \frac{2.51 \Delta x}{u}
$$

```python
nx =500
nt =500


u = 0.25 
alpha = 0 


t = np.linspace(0, 2, int(nt+1))
x = np.linspace(0, 1, int(nx+1))

dx = x[1] - x[0]
dt = t[1] - t[0]
```

```python
print(dt)
```

```python
print(dx)
```

```python
print("RHS = ", 2.51*dx/u)
```

```python
u*dt/dx
```

$$ \frac{u\Delta t}{\Delta x} < 2.51$$

```python
def f(y):
    # initialize
    y_next = np.zeros(len(y))
    
    # solve for spatial discretization
    for jj in range(len(y)):
        if jj == len(y) -1:
            # enforce periodicity
            term1 =  -u * (y[0] - y[jj - 1])/(2*dx)
            term2 = alpha* (y[0] - 2*y[jj] + y[jj - 1])/(dx**2)
        
        elif jj == 0:
            # enforce periodicity
            term1 = -u * (y[jj + 1] - y[-1])/(2*dx)
            term2 = alpha* (y[jj + 1] - 2*y[jj] + y[-1])/(dx**2)
            
        else:
            term1 = -u * (y[jj + 1] - y[jj - 1])/(2*dx)
            term2 = alpha* (y[jj + 1] - 2*y[jj] + y[jj - 1])/(dx**2)
            
        y_next[jj] = term1 + term2 
    return y_next
```

```python
def rk3(y, dt):
    f1 = f(y)
    f2 = f(y + 8/15*dt*f1)
    f3 = f(y + 1/4*dt*f1 + 5/12*dt*f2)
    return y + dt*(f1/4 + 3*f3/4)
```

```python
res = np.zeros((len(x), len(t)))

# initial condition
res[:, 0] = np.sin(np.pi*x)*(np.cos(5*np.pi*x) + np.sin(20*np.pi*x))

# propagte the solution using numerical methods: central FDA for spatial and RK3 for time. 
for ii in range(1, len(t)):
    res[:, ii] = rk3(res[:, ii-1], dt)
```

```python
nt1 = np.argmin(np.abs(t - 0.5))
nt2 = np.argmin(np.abs(t - 1))
nt3 = np.argmin(np.abs(t - 1.5))
```

```python
fig, ax = plt.subplots(ncols=2, nrows=2, sharex=True, sharey=True, figsize=(10, 5))

f_interp = interp1d(x, res[:, 0], fill_value="extrapolate", bounds_error=False)

_ = ax[0, 0].plot(x, res[:, 0])
_ = ax[0, 0].set_title("t=0")


_ = ax[0, 1].plot(x, res[:, nt1], label="numerical")
_ = ax[0, 1].set_title("t=0.5")
# define the new shifted coordinate.
x_new = (x - u*t[nt1]) % (1)
# interpolate over the shifted coordinate
_ = ax[0, 1].plot(x, f_interp(x_new), ls = "--", label="exact")


_ = ax[1, 0].plot(x, res[:, nt2], label="numerical")
# define the new shifted coordinate.
x_new = (x - u*t[nt2]) % (1)
# interpolate over the shifted coordinate
_ = ax[1, 0].plot(x, f_interp(x_new), ls = "--", label="exact")
_ = ax[1, 0].set_title("t=1")

_ = ax[1, 1].plot(x, res[:, -1], label="numerical")
# define the new shifted coordinate.
x_new = (x - u*t[-1]) % (1)
# interpolate over the shifted coordinate
_ = ax[1, 1].plot(x, f_interp(x_new), ls = "--", label="exact")
_ = ax[1, 1].set_title("t=2")


ax[0, 1].legend()
ax[1, 0].legend()
ax[1, 1].legend()

ax[1, 0].set_xlabel("x")
ax[1, 1].set_xlabel("x")
ax[1, 0].set_ylabel("T")
ax[0, 0].set_ylabel("T")

plt.savefig("figs3/advection_4plate.png", dpi=500)
```

```python
fig, ax = plt.subplots(sharex=True, sharey=True, figsize=(10, 5))

f_interp = interp1d(x, res[:, 0], fill_value="extrapolate", bounds_error=False)

# define the new shifted coordinate.
x_new = (x - u*t[nt1]) % (1)
# interpolate over the shifted coordinate
_ = ax.plot(x, np.abs(res[:, nt1] - f_interp(x_new)), ls = "--", label="t=0.5")

# define the new shifted coordinate.
x_new = (x - u*t[nt2]) % (1)
# interpolate over the shifted coordinate
_ = ax.plot(x, np.abs(res[:, nt2] - f_interp(x_new)), ls = "--", label="t=1")

x_new = (x - u*t[-1]) % (1)
# interpolate over the shifted coordinate
_ = ax.plot(x, np.abs(res[:, -1] - f_interp(x_new)), ls = "--", label="t=2")


ax.legend()

ax.set_xlabel("x")
ax.set_ylabel("Abs Error")
ax.set_yscale("log")

plt.savefig("figs3/advection_4plate_error.png", dpi=500)
```

# (b)
$$
 \frac{dT_{j}}{dt} = \alpha \frac{T_{j+1} - 2T_{j} + T_{j-1}}{(\Delta x)^2}
$$

CFL Condition: 

$$
\Delta t  < 0.6275 \frac{(\Delta x)^2}{\alpha}
$$

```python
nx =250
nt =1e4


u = 0
alpha = 0.005


t = np.linspace(0, 10, int(nt+1))
x = np.linspace(0, 1, int(nx+1))

dx = x[1] - x[0]
dt = t[1] - t[0]
```

```python
print("dt = ", dt)
```

```python
print("dx = ", dx)
```

```python
print("RHS =", 0.6275* ((dx)**2)/alpha)
```

```python
print(4*alpha*dt/(dx**2))
```

```python
res = np.zeros((len(x), len(t)))

# initial condition
res[:, 0] = np.sin(np.pi*x)*(np.cos(5*np.pi*x) + np.sin(20*np.pi*x))

# propagte the solution using numerical methods: central FDA for spatial and RK3 for time. 
for ii in range(1, len(t)):
    res[:, ii] = rk3(res[:, ii-1], dt)
```

```python
nt1 = np.argmin(np.abs(t - 0.025))
nt2 = np.argmin(np.abs(t - 0.1))
nt3 = np.argmin(np.abs(t - 0.5))
nt4 = np.argmin(np.abs(t - 1))
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

_ = ax.plot(x, res[:, 0], label="t=0 s")
_ = ax.plot(x, res[:, nt1], label="t=0.025 s")
_ = ax.plot(x, res[:, nt2], label="t=0.1 s")
_ = ax.plot(x, res[:, nt3], label="t=0.5 s")
_ = ax.plot(x, res[:, nt4], label="t=1 s")
_ = ax.plot(x, res[:, -1], label="t=10 s")

_ = ax.set_xlabel("x")
_ = ax.set_ylabel("T")
_ = ax.set_title("Diffusion")
ax.legend()


plt.savefig("figs3/diffusion_1plate.png", dpi=500)
```
