```python
h1 = 0.01 
h2 = 0.08
h3 = 1

tf = 500
```

```python
N1 = int(tf/h1) + 1
N2 = int(tf/h2) + 1
N3 = int(tf/h3) + 1
```

```python
t1 = np.linspace(0, tf, N1)
t2 = np.linspace(0, tf, N2)
t3 = np.linspace(0, tf, N3)
```

```python
omega = 0.2
tau = 400

def f(y, omega, tau):
    return complex(-1/tau, omega)*y

def y_next(y, omega, tau, theta, h):
    sigma = 1 + (h*complex(-1/tau, omega))/(1 - h*complex(-1/tau, omega)*theta)
    return (sigma)*y
    #return (y + h*(1-theta)*f(y, omega, tau))/(1 - h*theta*complex(-1/tau, omega))

def y_exact(t, omega, tau):
    return np.exp(complex(-1/tau, omega)*t)
```

```python
t1_res = np.zeros(len(t1), dtype = np.complex128)
t2_res = np.zeros(len(t2), dtype = np.complex128)
t3_res = np.zeros(len(t3), dtype = np.complex128)

t1_res[0] = 1
t2_res[0] = 1
t3_res[0] = 1

t1_exact = y_exact(t=t1, omega=omega, tau=tau)
t2_exact = y_exact(t=t2, omega=omega, tau=tau)
t3_exact = y_exact(t=t3, omega=omega, tau=tau)
```

```python
theta = 1/4

for ii in range(N1 -1):
    t1_res[ii + 1] = y_next(y=t1_res[ii], omega=omega, tau=tau, theta=theta, h=h1)
    
for ii in range(N2 -1):
    t2_res[ii + 1] = y_next(y=t2_res[ii], omega=omega, tau=tau, theta=theta, h=h2)
    
for ii in range(N3 -1):
    t3_res[ii + 1] = y_next(y=t3_res[ii], omega=omega, tau=tau, theta=theta, h=h3)
```

```python
t1_res_2 = np.zeros(len(t1), dtype = np.complex128)
t2_res_2 = np.zeros(len(t2), dtype = np.complex128)
t3_res_2 = np.zeros(len(t3), dtype = np.complex128)

t1_res_2[0] = 1
t2_res_2[0] = 1
t3_res_2[0] = 1
```

```python
theta = 4/5
for ii in range(N1 -1):
    t1_res_2[ii + 1] = y_next(y=t1_res_2[ii], omega=omega, tau=tau, theta=theta, h=h1)
    
for ii in range(N2 -1):
    t2_res_2[ii + 1] = y_next(y=t2_res_2[ii], omega=omega, tau=tau, theta=theta, h=h2)
    
for ii in range(N3 -1):
    t3_res_2[ii + 1] = y_next(y=t3_res_2[ii], omega=omega, tau=tau, theta=theta, h=h3)
```

```python
fig, ax = plt.subplots(ncols = 2, nrows=2, sharex = True, figsize=(10, 8))
ax[0, 0].plot(t1, np.abs(t1_res_2), ls = "--", label="h = 0.01 s")
ax[0, 0].plot(t2, np.abs(t2_res_2), ls = "-.",  label="h = 0.08 s")
ax[0, 0].plot(t3, np.abs(t3_res_2), ls = ":", label="h = 1 s")
ax[0, 0].plot(t3, np.abs(t3_exact), label="Exact")

ax[0, 1].plot(t1, np.abs(t1_res), ls = "--", label="h = 0.01 s")
ax[0, 1].plot(t2, np.abs(t2_res), ls = "-.",  label="h = 0.08 s")
ax[0, 1].plot(t3, np.abs(t3_res), ls = ":", label="h = 1 s")
ax[0, 1].plot(t3, np.abs(t3_exact), label="Exact")


ax[1, 1].plot(t1, np.arctan2(t1_res.imag, t1_res.real), ls = "--", label="h = 0.01 s")
ax[1, 1].plot(t2, np.arctan2(t2_res.imag, t2_res.real), ls = "-.",  label="h = 0.08 s")
ax[1, 1].plot(t3, np.arctan2(t3_res.imag, t3_res.real), ls = ":", label="h = 1 s")
ax[1, 1].plot(t3, np.arctan2(t3_exact.imag, t3_exact.real), label="Exact")


ax[1, 0].plot(t1, np.arctan2(t1_res_2.imag, t1_res_2.real), ls = "--", label="h = 0.01 s")
ax[1, 0].plot(t2, np.arctan2(t2_res_2.imag, t2_res_2.real), ls = "-.",  label="h = 0.08 s")
ax[1, 0].plot(t3, np.arctan2(t3_res_2.imag, t3_res_2.real), ls = ":", label="h = 1 s")
ax[1, 0].plot(t3, np.arctan2(t3_exact.imag, t3_exact.real), label="Exact")

ax[0, 1].set_ylim(0, 1.5)
ax[0, 0].set_title(r"$\theta = 4/5$" + " \n Amplitude")
ax[0, 1].set_title(r"$\theta = 1/4$" + " \n Amplitude")

ax[1, 0].set_title(r"$\theta = 4/5$" + " \n Phase")
ax[1, 1].set_title(r"$\theta = 1/4$" + " \n Phase")
ax[1, 0].set_xlabel(r"$t$")
ax[1, 1].set_xlabel(r"$t$")

ax[0,0].legend()
ax[0,1].legend()
ax[1,1].legend()
ax[1,0].legend()

ax[0, 0].set_xlim(0, 50)

plt.tight_layout()
plt.savefig("figs/amplitude_phase_error_50__.png")
```

# Problem 3

```python
ih = np.linspace(-3, 3, int(1000))
rh = np.linspace(-3, 3, int(1000))

res = np.ones((len(ih), len(rh)))
fig, ax = plt.subplots(figsize=(5, 5))

for inum, ii in enumerate(rh):
    for jnum, jj in enumerate(ih):
        lam = complex(ii, jj)
        expression = 1 + lam + (lam**2)/2 + (lam**3)/6
        if np.abs(expression) <=1:
            res[jnum , inum] = 0
            
plt.imshow(res, cmap="gray", extent=[-3,3,-3,3], aspect="auto")
            
ax.set_title("Stability region for 3rd-order RK")
ax.set_xlabel(r"$Re(\lambda h)$")
ax.set_ylabel(r"$Im(\lambda h)$")

plt.savefig("figs/rk3_stability")
```

```python
def f(y, omega, tau):
    return complex(-1/tau, omega)*y
```

```python
def y_next_rk3(y, omega, tau, h):
    f1 = f(y=y, omega=omega, tau=tau)
    f2 = f(y=(y + (8/15)*h*f1), omega=omega, tau=tau)
    f3 = f(y=(y + (1/4)*h*f1 + (5/12)*h*f2), omega=omega, tau=tau)
    return y + h*(f1/4 + (3/4)*f3)
```

```python
rk3_res_1 = np.zeros(N1, dtype = 'complex_')
rk3_res_2 = np.zeros(N2, dtype = 'complex_')
rk3_res_3 = np.zeros(N3, dtype = 'complex_')

rk3_res_1[0] = 1
rk3_res_2[0] = 1
rk3_res_3[0] = 1
```

```python
for ii in range(N1 -1):
    rk3_res_1[ii + 1] = y_next_rk3(y=rk3_res_1[ii], omega=omega, tau=tau, h=0.01)
    
for ii in range(N2 -1):
    rk3_res_2[ii + 1] = y_next_rk3(y=rk3_res_2[ii], omega=omega, tau=tau, h=0.08)
    
for ii in range(N3 -1):
    rk3_res_3[ii + 1] = y_next_rk3(y=rk3_res_3[ii], omega=omega, tau=tau, h=1)
```

```python
lambda_curr = complex(-1/tau, omega)
```

```python
 np.abs(1 + lambda_curr + 1/2 * lambda_curr**2 +  1/6 * lambda_curr**3)
```

```python
fig, ax = plt.subplots(ncols = 2, figsize=(10, 5))

ax[0].plot(t3, np.abs(rk3_res_3), ls="-", label="h = 1s")
ax[0].plot(t2, np.abs(rk3_res_2), ls="-", label="h = 0.08s")
ax[0].plot(t1, np.abs(rk3_res_1), ls="-", label="h = 0.01s")

ax[0].plot(t3, np.abs(t3_exact), ls="-", label="Exact")


ax[1].plot(t3, np.abs(np.abs(rk3_res_3) - np.abs(t3_exact)), ls="-", label="h = 1s")
ax[1].plot(t2, np.abs(np.abs(rk3_res_2) - np.abs(t2_exact)), ls="-", label="h = 0.08s")
ax[1].plot(t1, np.abs(np.abs(rk3_res_1) - np.abs(t1_exact)), ls="-", label="h = 0.01s")

ax[1].set_xlabel("t")
ax[0].set_xlabel("t")
ax[0].set_ylabel("Amplitude")
ax[1].set_ylabel("Amplitude Abs Error")
ax[1].legend()
ax[0].legend()
ax[1].set_yscale("log")
plt.tight_layout()

plt.savefig("figs/rk3_amp_error.png")
```
